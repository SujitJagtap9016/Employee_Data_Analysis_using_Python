```python
cities = df['City'].unique()
fig, axes = plt.subplots(1, len(cities), figsize=(15, 5)) 

for i, city in enumerate(cities):
    ax = sns.countplot(x='Gender', hue='Gender', data=df[df['City'] == city], ax=axes[i])
    ax.set_title(f'Gender Distribution in {city}')

    for container in ax.containers: 
        ax.bar_label(container) 

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Age', hue='LeaveOrNot', data=df)

for container in ax.containers: 
        ax.bar_label(container) 

plt.title('Employee Attrition by Age')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.legend(title='Left Company')

plt.show()
```

```python
city_gender_counts = df.groupby(['City', 'Gender']).size().unstack()
city_gender_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Proportion of Employees by City and Gender')
plt.xlabel('City')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='City', hue='Gender', data=df)
plt.title('Employee Distribution by City and Gender')
plt.xlabel('City')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.legend(title='Gender')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                xy=(p.get_x() + p.get_width() / 2, height), 
                ha='center', va='bottom')

plt.show()
```

```python
df.head()
```

```python
city_leave_counts = df.groupby(['City', 'LeaveOrNot']).size().unstack()
city_leave_counts = city_leave_counts.div(city_leave_counts.sum(axis=1), axis=0)

for city in city_leave_counts.index:
    plt.figure(figsize=(5, 5))
    plt.pie(city_leave_counts.loc[city], labels=city_leave_counts.columns, autopct='%1.1f%%', startangle=90)
    plt.title(f'Leave Status in {city}')
    plt.show()
```

```python
g = sns.FacetGrid(df, col="City", row="Gender")
g.map(sns.histplot, "Age")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x='City', y='Age', hue='Education', data=df)
plt.title('Average Age by City and Education')
plt.xlabel('City')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.legend(title='Education')

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, height + 5, f'{int(height)}', 
                ha='center', va='bottom')

plt.show()
```

```python
df_stayed = df[df['LeaveOrNot'] == 'No']
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=df_stayed, x='Age', bins=20) 
ax.bar_label(ax.containers[0]) 
plt.title('Age Distribution of Employees Who Stayed')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Employee Distribution by City and Gender (by Education)')

for i, edu in enumerate(df['Education'].unique()):
    ax = sns.countplot(x='City', hue='Gender', data=df[df['Education'] == edu], ax=axes[i])
    for container in ax.containers: 
        ax.bar_label(container) 
    axes[i].set_title(f'Education: {edu}')
    axes[i].set_xlabel('City')
    axes[i].set_ylabel('Number of Employees')
    axes[i].legend(title='Gender')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
